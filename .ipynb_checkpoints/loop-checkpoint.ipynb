```python
while True:
    num1=float(input("enter your first number:"))
    op=input("enter operator(+, _, *, /):")
    num2=float(input("enter your second number:"))
    result=0
    if op== '+':
        result = num1 + num2
    elif op== '-':
        result = num1 - num2
    elif op== '*':
        result = num1 * num2
    elif op== '/':
        result = num1 / num2
    print(f"result is: {result}")
  
choice = input("type 'yes' to continue or 'quite' to exit: ")
if choice.lower() == 'quit':
    break
```

```python
import random

x=20
magic_number = random.randint(1, x)
attempt = 0

while True:
    guess = int(input(f"enter a number between 1 and {x}:"))
    attempt += 1
    if guess==magic_number:
        print(f"you guess the number is correct {attempt} attempt.")
        break
    elif guess < magic_number:
        print("the entered number is smaller than the magic number.")
    else:
        print("the entered number is larger than the magic number.")
        break
        
        
```

```python
import random
rock=1
paper=2
scissors=3

user=int(input("enter your choice(1.rock,2.paper,3.scissors):"))
computer=random.randint(1,3)

if user==rock and computer==scissors:
    print("rock beat scissors")
elif user==paper and computer==rock:
    print("paper beats rock")
elif user==scissors and computer==paper:
    print("scissors beat paper")
elif user==computer:
    print("it is tie")
else:
    print("you lose")
```

```python
import random
n=3
user_count=0
ai_count=0
while True:
    rock=1
    paper=2
    scissor=3
    user=int(input("enter your choice(1.rock,2.paper,3.scissor):"))
    ai=random.randint(1,3)

if(user==rock and ai==paper):
    user_count=user_count+1
elif user==paper and ai==scissor:
    ai_count=ai_count+1
    
if(user_count>ai_count):
    print("user win")
elif (ai_count>user_count):
    print("ai win")
elif (user_count==ai_count):
    print("draw")
else:
    print("lose")
    break
    
```
