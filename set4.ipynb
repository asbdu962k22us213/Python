```python
def graph():
    x=int(input("enter a value of x: "))
    y=int(input("enter a value of y: "))
    if(x>0 and y>0):
        return ("first quadrant")
    if(x>0 and y<0):
        return ("second quadrant")
    if(x<0 and y<0):
        return ("third quadrant")
    if(x<0 and y>0):
        return ("fourth quadrant")
    if(x==0 and y!=0):
        return ("fifth quadrant")
    if(x!=0 and y==0):
        return ("sixth quadrant")
    if(x==0 and y==0):
        return ("seventh quadrant")
    else:
        return ("invalid")
graph()
    
```

```python
def simple(num1,num2):
    op= input('Input an operator(+, -, *, /, %):')
    if op=='+':
        add=num1 + num2
        return (f"result: {add}")
    elif op=='-':
        sub=num1 - num2
        return (f"result: {sub}")
    elif op=='*':
        mul=num1 * num2
        return (f"result: {mul}")
    elif op=='/':
        div=num1 / num2
        return (f"result: {div}")
    elif op=='%':
        mod=num1 % num2
        return (f"result: {mod}")
    else:
        return ("invalid")

num1= float(input("enter a value:"))
num2= float(input("enter a value:"))
simple(num1,num2)
```
