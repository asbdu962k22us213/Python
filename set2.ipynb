```python
num1= float(input("enter a value:"))
num2= float(input("enter a value:"))
op= input('Input an operator(+, -, *, /, %):')
if op=='+':
    add=num1 + num2
    print(f"result: {add}")
elif op=='-':
    sub=num1 - num2
    print(f"result: {sub}")
elif op=='*':
    mul=num1 * num2
    print(f"result: {mul}")
elif op=='/':
    div=num1 / num2
    print(f"result: {div}")
elif op=='%':
    mod=num1 % num2
    print(f"result: {mod}")
else:
    print("invalid")
```
